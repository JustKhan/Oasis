```python
cloth=df[df['category']=='Clothing']
print(cloth)
clothprice_mean= cloth[['price']].mean()
print(clothprice_mean)
```

```python
clothmale=cloth[cloth['gender']=='Male']
print(clothmale)
```

```python
clothmale_mean= clothmale[['price']].mean(axis=0)
clothmale_sum= clothmale[['price']].sum()
print("clothmale_mean is:",clothmale_mean)
print("clothmale_sum is:", clothmale_sum)
```

```python
clothfemale=cloth[cloth['gender']=='Female']
print(clothfemale)
clothfemale_mean= clothfemale[['price']].mean()
print("clothfemale_mean is:", clothfemale_mean)
clothfemale_sum= clothfemale[['price']].sum()
print("clothfemale_sum is:", clothfemale_sum)

a=float(clothmale_sum)
b=float(clothfemale_sum)
print(a)
print(b)
```

```python
x_axis=["Clothe_male_sum","Clothe_Female_sum"]
y_axis=[a,b]

plt.bar(x_axis,y_axis)
plt.show()
print("Cloth male sum is",a)
print("Clothe female sum is",b)
```

Thus We can see that more sales has been done by Female


```python
cloth_age_mean= cloth[['age']].mean()
print("Mean age of Cloth customers is:",cloth_age_mean)

cloth_male_age_mean=clothmale[['age']].mean()
print("Mean age of Male clothing customers is: ",cloth_male_age_mean)

cloth_female_age_mean=clothfemale[['age']].mean()
print("Mean age of female clothing customers is: ",cloth_female_age_mean)
```

```python
ages_buying_books = df[df['category'] == 'Books'].groupby('age')['age'].value_counts().sort_values(ascending= False)
ages_buying_books
```

```python
fig = px.bar(ages_buying_books,
            template='plotly_dark',
            title = 'Different ages buying books',
             color=ages_buying_books.values,
             y=ages_buying_books.values,
             x=ages_buying_books.index,
             labels={'y':'Number of Books'}
            )
fig.show()
```
